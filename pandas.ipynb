## Fun with pandas

```python
#load your data into pandas

import pandas as pd

df = pd.read_csv('pokemon_data.csv')

print(df.head())
```

### Reading our data in pandas

```python
# read headers/column names

print(df.columns)
```

```python
# Read each 'name' in the column

print(df['Name'][0:10]) 

print(df[['Name','Type 1']][0:10])
```

```python
# read each row 

print(df.iloc[1:5])
```

```python
# read a specific location

print(df.iloc[4,1])
```

```python
# using iterrows()

for index,row in df.iterrows():
    print(index,row['Name'])
```

```python
# using .loc 

df.loc[df['Type 1']=='Water'].head()
```

### Sorting and describing our data

```python
df.describe()
```

```python
# sort by 'Name'

df.sort_values('Name').head()
```

```python
# sort by 'Name', ascending order

df.sort_values('Name', ascending=False).head()
```

```python
# more columns

df.sort_values(['Type 1','HP']).head()
```

### Changing our data

```python
# adding a column

df['Total']=df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
print(df.head())
```

```python
# drop a column

df=df.drop(columns=['Total'])
print(df.head())
```

```python
# add a column using iloc 

df['Total']=df.iloc[:,4:10].sum(axis=1)
df.head(5)
```

### Saving our data 

```python
# save as .csv

df.to_csv('modified.csv', index=False)
```

```python
# save as .xlsx

df.to_excel('modified.xlsx', index=False)
```

### Filtering 

```python
# two or more conditions

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2']=='Poison') & (df['HP'] > 70 )]
```

### Conditional changes


```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df.tail()
```

```python
df.loc[df['Type 1']=='Flamer', 'Type 1'] = 'Fire'
df.tail()
```

### Groupby

```python
# groupby 'Type 1' using .mean()

df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
# groupby 'Type 1' using .count()

df['count']=1

df.groupby(['Type 1']).count()['count']
```

### Working with large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize=5):
    print("Chunk dataframe")
    print(df)
```
